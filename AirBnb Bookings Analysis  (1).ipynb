# 1.How does the price of Airbnb listings vary by location (city or neighborhood)?

```python
Price_Shift=new_air.groupby('neighbourhood_group')['price'].var().sort_values(ascending=False).round(4).rename({'neighbourhood_group':'Location'})
Price_Shift
```

```python
plt.scatter(new_air['neighbourhood_group'],new_air['price'],color='black')
plt.xlabel('neighbourhood_group')
plt.ylabel('price')
plt.title('Price Shift')
plt.show()
```

# Conclusion
Queens shows the greatest variation among all the other locations, whereas Bronz shows the least variation in the price listings




# 2.Which is the most preferred location?



```python
#using value counts function to get the most preferred location
perferred_location=new_air.neighbourhood_group.value_counts().sort_values(ascending=False)
pd.DataFrame(perferred_location)
```

```python
#pie chart to enhance the understanding of the above data
plt.pie(perferred_location, labels=perferred_location.index, shadow = True)
plt.title("Preferability as per the location")
plt.show()
```

# Conclusion

Brooklyn is the most preferred location and Staten Island being the least preferred.



# 3.What is the average preferred price by customers according to the location?


```python
#grouping by location and aggregating the price to get the mean
avg_preferred_price=new_air.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False).round(2).rename({'neighbourhood_group':'Location','price':'Average Price'})
pd.DataFrame(avg_preferred_price)
```

```python
#plotting a bar graph for a better understanding
plt.bar(avg_preferred_price.index,avg_preferred_price.values,color = "black")
plt.xlabel("Average preferred price")
plt.ylabel("Location")
plt.title("Average preferred price by customers according to the location")
plt.show()
```

# Conclusion

 Mahattan has the highest average price, followed by Brooklyn.
 Bronx has the least average pric.





# 4.Most demanded host as per the location



```python
# Find the host with the maximum count of listings
host_count=new_air['host_name'].value_counts()
host_counts=pd.DataFrame({'Host_name':host_count.index,'listing_count':host_count.values})
sorting_values=host_counts.sort_values(by='listing_count',ascending=False)
sorting_values.head()
```

```python
sorting_values=sorting_values.head(10)
# Define a list of custom colors for each bar
custom_colors = ['crimson', 'navy', 'lime', 'orchid', 'gold', 'teal', 'coral', 'violet', 'sienna', 'olive']
# Creating a bar plot with custom colors for each bar
plt.figure(figsize=(10, 6))
bars = plt.bar(sorting_values['Host_name'].astype(str), sorting_values['listing_count'], color=custom_colors)
plt.xlabel('Host name')
plt.ylabel('Number of Listings')
plt.title('Top Hosts by Number of Listings')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.legend(bars, sorting_values['Host_name'].astype(str), title='Host name')
plt.tight_layout()
plt.show()
```

# Conclusion
Michael is the most demanded host, followed by David and John.



# 5.Which neighborhoods have the most Airbnb listings within each neighborhood group?



```python
# Group by "neighbourhood_group" and then count listings in each "neighbourhood" within the group
neighbourhood=new_air.groupby('neighbourhood')['neighbourhood_group'].value_counts().reset_index(name='listing_count')
neighbourhood=neighbourhood.sort_values(by='listing_count',ascending=False)
pd.DataFrame(neighbourhood)
```

```python
top_neighbourhood = neighbourhood.groupby("neighbourhood_group").head(5)

# Pivot the DataFrame for easier plotting
pivot_df = top_neighbourhood.pivot(index="neighbourhood", columns="neighbourhood_group", values="listing_count")

# Create a horizontal bar chart using Pandas plot
ax = pivot_df.plot(kind="bar", figsize=(10, 6), color=plt.cm.viridis(np.linspace(0, 1, 5)))
ax.set_xlabel('Number of Listings')
ax.set_ylabel('Neighborhood')
ax.set_title('Top Neighborhoods by Number of Listings within Each Group')
ax.legend(title='Neighborhood Group', loc='upper right')

plt.tight_layout()
plt.show()
```

# 6.What is the maximum and minimum price at a location?



```python
#getting the highest prices as per the location by grouping by on the basis of location
highest_price=(new_air.groupby('neighbourhood_group',as_index=False)['price'].max().sort_values(['price'],ascending=False).rename(columns = {'price':'Highest price','neighbourhood_group':'Location'}))
highest_price
```

```python
#getting the lowest prices as per the location by grouping by on the basis of location
lowest_price=(new_air.groupby('neighbourhood_group',as_index=False)['price'].min().sort_values(['price'],ascending=True).rename(columns = {'price':'Highest price','neighbourhood_group':'Location'}))
lowest_price
```

```python
prices_merge=pd.merge(highest_price,lowest_price,on="Location")
prices_merge
```

# Conclusion
Brooklyn offers the maximum price, whereas Bronx offers the least price.



# 7.Which roomtype is most referred by hosts ?



```python
# The code you provided will return a DataFrame of the top room type preferred by the maximum number of hosts.
referred_roomtype=new_air.groupby('room_type')['host_name'].count().sort_values(ascending=False).reset_index()
pd.DataFrame(referred_roomtype)
```

```python
explode_values=(0,0,0.2)      #specifying the explode values for each room type in pie chart
plt.pie(referred_roomtype['host_name'],labels=referred_roomtype['room_type'],shadow=True,explode=explode_values)
plt.title("The most preferred room type by hosts")
plt.legend()
plt.show()
     
```

# Conclusion
Entire home/apt is the most referred room type by the hosts, shared rooms being the least referred.



# 8.Comparing the availabilty of hosts



```python
#hosts that are not available for the whole year i.e., with zero availability
host_availability_0=new_air[new_air['availability_365']==0][['host_name','availability_365']]
host_availability_0
```

```python
#hosts with zero availability (top 5)
total_host_availability_0=host_availability_0.groupby('host_name')['availability_365'].count().sort_values(ascending=False)
top5_hosts=pd.DataFrame(total_host_availability_0.head())
top5_hosts
```

```python
colors = ['mediumslateblue', 'mediumaquamarine', 'mediumpurple', 'mediumturquoise', 'mediumspringgreen']
# Create a bar chart with different colors for each bar
plt.figure(figsize=(10, 6))
for i in range(len(top5_hosts)):
    plt.bar(top5_hosts.index[i], top5_hosts.values[i], color=colors[i])

plt.xlabel("Host Name")
plt.ylabel("Number of Availability Values")
plt.title("Top 5 Hosts with Most Availability")
plt.xticks(rotation=45)
plt.show()
```

# Conclusion
Michael is the least available host,with the highest zero availability counts




```python
#hosts available for more than half of the year,i.e., availability>183 days of the year
host_availability_more_than_halfyear=new_air[new_air['availability_365']>183][['host_name','availability_365']]
host_availability_more_than_halfyear
```

```python
#hosts with the most availabilty (top 5)
total_host_availability_more_than_halfyear=host_availability_more_than_halfyear.groupby('host_name')['availability_365'].count().sort_values(ascending=False)
top5_hosts=pd.DataFrame(total_host_availability_more_than_halfyear.head())
top5_hosts
```

```python
colors = ['gold', 'darkorange', 'crimson', 'darkorchid', 'steelblue']
# Create a bar chart with different colors for each bar
plt.figure(figsize=(10, 6))
for i in range(len(top5_hosts)):
    plt.bar(top5_hosts.index[i], top5_hosts.values[i], color=colors[i])

plt.xlabel("Host Name")
plt.ylabel("Number of Availability Values")
plt.title("Top 5 Hosts with Most Availability")
plt.xticks(rotation=45)
plt.show()
```

# Conclusion
Sonder is the most available host,with the maximum availability counts



```python
#merging the dataframe of hosts with zero avilability and with more than 183 days of availability
merged=pd.merge(host_availability_0,host_availability_more_than_halfyear,how="outer")
merged
```

# 9.Which host offers the economical prices (between 100 and 250)?



```python
economical_price=new_air[(new_air['price']>=100 )& (new_air['price']<=250)][['host_name','price']].sort_values(by='price',ascending=False)
economical_price
```

```python
#which host offers the most economical prices (top 5)
host_offering_economical_prices=economical_price.groupby('host_name')['price'].count().sort_values(ascending=False)
host_offering_economical_prices.head()
```

```python
#visual representation of the top 5 hosts offering economical prices
plt.figure(figsize=(10,6))
bars=plt.bar(host_offering_economical_prices.head().index,host_offering_economical_prices.head().values,color='blue')
plt.xlabel("Hosts Name")
plt.ylabel("Counts of Economical Prices")
plt.title("Top 5 Hosts Offering the Most Economical Prices")
plt.show()
```

# Conclusion
Michael offers most of the economical prices



# 10.What are the top 10 neighborhoods with the highest average nightly prices for accommodations, as depicted in the 'Average Nightly Price Neighborhood' plot?



```python
average_prices_by_neighborhood=new_air.groupby('neighbourhood')['price'].mean().head(10).round(2)
average_prices_by_neighborhood
```

```python
plt.figure(figsize=(10,6))
plt.plot(average_prices_by_neighborhood.head(10).index,average_prices_by_neighborhood.head(10).values,marker='o',color='gold')
plt.xlabel('Neighborhood')
plt.ylabel('Average Nightly Price')
plt.title('Average Nightly Price by Neighborhood')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

# Conclusion

Varied Neighborhood Prices: Prices vary significantly among the top 10 neighborhoods.

Budget Consideration: Travelers can use this graph to consider budget-friendly options when choosing a neighborhood to stay in.
